```python
import pandas as pd
```

```python
data=pd.read_csv('/content/3campain.csv')
```

```python
data
```

```python
data['Sell'].mean()
```

```python
data.groupby('Campain').mean()
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
res=ols('Sell ~ Campain',data=data).fit()
```

```python
sm.stats.anova_lm(res)
```
