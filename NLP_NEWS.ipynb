```python
df['Category2'] = df['Category2'].str.replace('\n', '')
```

```python
df['title_body_tokenized']=df['title_body'].apply(lambda x: word_tokenize(str(x)))
```

```python
df
```

```python
import nltk 
from nltk.corpus import stopwords
from string import punctuation
```

```python
nltk.download('stopwords')
```

```python
with open('stopwords.txt', 'r', encoding='utf-8') as stopwords_file:
    stopwords = stopwords_file.readlines()
stopwords = [line.replace('\n', '') for line in stopwords]
stopwords
```

```python
df['title_body'] = df['title_body'].replace('\u200c', '')
```

```python
df
```

```python
len(stopwords)
```

```python
nltk_stopwords = nltk.corpus.stopwords.words('english')
stopwords.extend(nltk_stopwords)
len(stopwords)
```

```python
from hazm import word_tokenize
from hazm import Stemmer
st=Stemmer()
```

```python
corpus=[]
for w in range(len(df['title_body'])) :
    msg=df['title_body'][w]
    msg=word_tokenize(msg)
    msg=[st.stem(word) for word in msg if not word in stopwords]
    msg=" ".join(msg)
    corpus.append(msg)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf=TfidfVectorizer()
X=tf.fit_transform(corpus).toarray()
```

```python
X.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(df['Category2'])
```

```python
len(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.svm import SVC
svm=SVC()
svm.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report ,confusion_matrix
y_pred=svm.predict(X_test)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
