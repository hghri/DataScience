```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/CFB2019.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
dfn=df['Win-Loss'].str.split('-',expand=True)
```

```python
df['Win']=dfn[0]
df['Loss']=dfn[1]
```

```python
df
```

```python
df['Win']=df['Win'].apply(pd.to_numeric,errors='coerce')
```

```python
df['Loss']=df['Loss'].apply(pd.to_numeric,errors='coerce')
```

```python
df_corr=df.corr()
```

```python
df_corr.sort_values(by='Win')
```

```python
df_corr[df_corr['Win']>0.70]['Win']
```

```python
df_n=df_corr[df_corr['Win']>0.70][['Games','Off Plays','Off Yards','Off TDs','First Downs','Redzone Attempts','Redzone Scores','Touchdowns','PAT','Total Points']]
plt.figure(figsize=(20,12))
sns.heatmap(df_n,annot=True)
```

```python
sns.pairplot(df_n)
```
