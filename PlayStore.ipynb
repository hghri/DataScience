```python
df=df.dropna(subset='Current Ver')
```

```python
df['Type'].unique()
```

```python
type_mapping={'Paid':1,'Free':0}
df['Type']=df['Type'].replace(type_mapping)
```

```python
df.groupby(['Type'])['Rating'].mean().plot(kind='bar' , figsize=(8,8))
```

```python
df['Size'] = df['Size'].replace('Varies with device', np.nan)
```

```python
def convert_size(val):
    if isinstance(val, str):
        if 'k' in val:
            return float(val.replace('k', '')) / 1000
        elif 'M' in val:
            return float(val.replace('M', ''))
        else:
            return None
    else:
        return val
```

```python
df['Size'] = df['Size'].apply(convert_size)
```

```python
df['Size'].mean()
```

```python
df['Size'].corr(df['Rating'])
```

```python
plt.scatter(df['Size'],df['Rating'],alpha=0.5)
plt.title('Relationship between Size and Rating')
plt.xlabel('Size')
plt.ylabel('Rating')
plt.show()
```

```python
df['Installs']=df['Installs'].str.replace('+' , '').str.replace(',','').astype(float)
```
